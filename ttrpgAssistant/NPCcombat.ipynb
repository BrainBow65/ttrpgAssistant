```python
from PySide6.QtWidgets import QApplication, QMainWindow, QPushButton, QVBoxLayout, QWidget, QComboBox
from PySide6.QtCore import Slot
import sys
import random
import json
```

```python
class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()

        self.setWindowTitle("NPC Generator")

        self.layout = QVBoxLayout()
        self.widget = QWidget()

        self.attributes = ['Strength', 'Dexterity', 'Constitution', 'Intelligence', 'Wisdom', 'Charisma']
        self.attribute_values = {attr: QComboBox() for attr in self.attributes}

        for attr, combo_box in self.attribute_values.items():
            combo_box.addItem('Random')
            for i in range(1, 21):
                combo_box.addItem(str(i))
            self.layout.addWidget(combo_box)

        self.generate_button = QPushButton("Generate NPC")
        self.generate_button.clicked.connect(self.generate_npc)

        self.layout.addWidget(self.generate_button)
        self.widget.setLayout(self.layout)

        self.setCentralWidget(self.widget)

    @Slot()
    def generate_npc(self):
        npc = {}
        for attr, combo_box in self.attribute_values.items():
            value = combo_box.currentText()
            if value == 'Random':
                value = random.randint(1, 20)
            npc[attr] = value
        print(npc)

app = QApplication([])
window = MainWindow()
window.show()

app.exec()
```

```python
import random

class Combatant:
    def __init__(self, name, hp, ac, attack_bonus, damage_bonus):
        self.name = name
        self.max_hp = hp
        self.hp = hp
        self.ac = ac
        self.attack_bonus = attack_bonus
        self.damage_bonus = damage_bonus

    def attack(self, target):
        roll = random.randint(1, 20) + self.attack_bonus
        if roll >= target.ac:
            damage = random.randint(1, 6) + self.damage_bonus
            target.hp -= damage
            print(f"{self.name} attacks {target.name} for {damage} damage.")
        else:
            print(f"{self.name}'s attack misses.")

    def heal(self, amount):
        self.hp += amount
        if self.hp > self.max_hp:
            self.hp = self.max_hp
        print(f"{self.name} heals for {amount} HP.")

class Encounter:
    def __init__(self, combatants):
        self.combatants = combatants

    def run(self):
        while True:
            for combatant in self.combatants:
                if combatant.hp <= 0:
                    print(f"{combatant.name} is defeated.")
                    self.combatants.remove(combatant)
                else:
                    target = random.choice([c for c in self.combatants if c != combatant])
                    combatant.attack(target)
            if len(self.combatants) <= 1:
                break
        print(f"{self.combatants[0].name} is the winner!")

# Example usage:
combatants = [
    Combatant("Player", 20, 15, 5, 3),
    Combatant("Enemy", 15, 12, 4, 2)
]
encounter = Encounter(combatants)
encounter.run()
```
